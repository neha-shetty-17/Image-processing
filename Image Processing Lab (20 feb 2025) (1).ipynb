```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread('dog.jpg')

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply median blur to the grayscale image
gray = cv2.medianBlur(gray, 5)

# Detect edges using the Laplacian method
edges = cv2.Laplacian(gray, cv2.CV_8U, ksize=5)
edges = 255 - edges

# Threshold the edges to create a binary mask
_, mask = cv2.threshold(edges, 100, 255, cv2.THRESH_BINARY)

# Apply bilateral filter to the original image
filtered = cv2.bilateralFilter(img, 9, 75, 75)

# Combine the filtered image with the edge mask
cartoon = cv2.bitwise_and(filtered, filtered, mask=mask)

# Convert BGR to RGB for matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cartoon_rgb = cv2.cvtColor(cartoon, cv2.COLOR_BGR2RGB)

# Display the original and cartoon images using matplotlib
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Cartoon Image')
plt.imshow(cartoon_rgb)
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('dog.jpg')

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)

# Perform Canny edge detection
edges = cv2.Canny(blurred_image, 100, 200)

# Display the original and processed images using matplotlib
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Edges Detected')
plt.imshow(edges, cmap='gray')
plt.axis('off')

plt.show()
```

```python
import cv2

# Load the image
image = cv2.imread('dog.jpg')

# Draw a rectangle
cv2.rectangle(image, (50, 50), (200, 200), (0, 255, 0), 3)

# Draw a circle
cv2.circle(image, (300, 300), 50, (255, 0, 0), -1)

# Draw a line
cv2.line(image, (100, 100), (400, 400), (0, 0, 255), 5)

# Draw text
cv2.putText(image, 'Hello, OpenCV!', (10, 500), cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 3)

# Display the image with shapes
cv2.imshow('Image with Shapes', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
