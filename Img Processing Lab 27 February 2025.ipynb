```python
import cv2
import matplotlib.pyplot as plt

# Load the image using OpenCV
image = cv2.imread('Nature.jpg')

# Convert the image from BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Get the dimensions of the image
height, width, _ = image_rgb.shape

# Define the function to split the image
def split_image(image_rgb):
    # Split the image into 4 parts
    top_left = image_rgb[:height // 2, :width // 2]
    top_right = image_rgb[:height // 2, width // 2:]
    bottom_left = image_rgb[height // 2:, :width // 2]
    bottom_right = image_rgb[height // 2:, width // 2:]

    return top_left, top_right, bottom_left, bottom_right

# Get the 4 parts of the image
top_left, top_right, bottom_left, bottom_right = split_image(image_rgb)

# Display the 4 parts using matplotlib
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

axes[0, 0].imshow(top_left)
axes[0, 0].set_title("Top Left")

axes[0, 1].imshow(top_right)
axes[0, 1].set_title("Top Right")

axes[1, 0].imshow(bottom_left)
axes[1, 0].set_title("Bottom Left")

axes[1, 1].imshow(bottom_right)
axes[1, 1].set_title("Bottom Right")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('Nature.jpg')
newimg = np.asarray(img)
X, Y, D = newimg.shape
div1 = X // 2
div2 = Y // 2

top_left = newimg[:div1, :div2]
top_right = newimg[:div1, div2:]
bottom_left = newimg[div1:, :div2]
bottom_right = newimg[div1:, div2:]

div_img = [top_left, top_right, bottom_left, bottom_right]
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

for idx, ax in enumerate(axs.flatten()):
    ax.imshow(cv2.cvtColor(div_img[idx], cv2.COLOR_BGR2RGB))
    ax.set_title(['Top Left', 'Top Right', 'Bottom Left', 'Bottom Right'][idx])
    ax.axis('off')

plt.show()
```
