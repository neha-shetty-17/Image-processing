```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image and convert to grayscale
image = cv2.imread('catt.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Define kernel
kernel = np.ones((5, 5), np.uint8)

# Apply erosion and dilation
erosion_edge = np.clip(cv2.subtract(gray_image, cv2.erode(gray_image, kernel)), 0, 255)
dilation_edge = np.clip(cv2.subtract(gray_image, cv2.dilate(gray_image, kernel)), 0, 255)

# Convert edge images to RGB for display
erosion_edge_rgb = cv2.cvtColor(erosion_edge.astype(np.uint8), cv2.COLOR_GRAY2BGR)
dilation_edge_rgb = cv2.cvtColor(dilation_edge.astype(np.uint8), cv2.COLOR_GRAY2BGR)

# Display results
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original Image'), plt.axis('off')
plt.subplot(1, 3, 2), plt.imshow(cv2.cvtColor(erosion_edge_rgb, cv2.COLOR_BGR2RGB)), plt.title('Edge after Erosion'), plt.axis('off')
plt.subplot(1, 3, 3), plt.imshow(cv2.cvtColor(dilation_edge_rgb, cv2.COLOR_BGR2RGB)), plt.title('Edge after Dilation'), plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
image_path = 'flower.jpg'  # Change this to your image path
image = cv2.imread(image_path)

if image is None:
    print("Error: Could not read the image.")
    exit()

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply erosion and dilation
kernel = np.ones((5, 5), np.uint8)
eroded = cv2.erode(gray, kernel)
dilated = cv2.dilate(gray, kernel)

# Edge images
edge_erosion = cv2.subtract(gray, eroded)
edge_dilation = cv2.subtract(gray, dilated)

# Display the results
plt.figure(figsize=(12, 8))
titles = ['Original', 'Eroded', 'Edge (Erosion)', 'Dilated', 'Edge (Dilation)']
images = [image, eroded, edge_erosion, dilated, edge_dilation]
for i in range(5):
    plt.subplot(2, 3, i + 1)
    plt.imshow(images[i], cmap='gray' if i != 0 else None)
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
