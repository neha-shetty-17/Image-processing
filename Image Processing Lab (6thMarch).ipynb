```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Read the input image
img = cv2.imread("city.jpg")
# Convert from BGR to RGB so we can plot using matplotlib
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Disable x & y axis and show the image
plt.axis('off')
plt.imshow(img)
plt.show()

# Get the image shape
rows, cols, dim = img.shape

# Transformation matrix for translation
M_translate = np.float32([[1, 0, 50],
                          [0, 1, 50],
                          [0, 0, 1]])
# Apply a perspective transformation to the image
translated_img = cv2.warpPerspective(img, M_translate, (cols, rows))

# Disable x & y axis and show the translated image
plt.axis('off')
plt.imshow(translated_img)
plt.show()

# Save the translated image to disk
plt.imsave("city_translated.jpg", translated_img)

# Transformation matrix for scaling
M_scale = np.float32([[1.5, 0, 0],
                      [0, 1.8, 0],
                      [0, 0, 1]])
# Apply a perspective transformation to the image
scaled_img = cv2.warpPerspective(img, M_scale, (cols * 2, rows * 2))

# Disable x & y axis and show the scaled image
plt.axis('off')
plt.imshow(scaled_img)
plt.show()

# Save the scaled image to disk
plt.imsave("city_scaled.jpg", scaled_img)

# Angle from degree to radian
angle = np.radians(10)
# Transformation matrix for rotation
M_rotate = np.float32([[np.cos(angle), -np.sin(angle), 0],
                       [np.sin(angle), np.cos(angle), 0],
                       [0, 0, 1]])
# Apply a perspective transformation to the image
rotated_img = cv2.warpPerspective(img, M_rotate, (int(cols), int(rows)))

# Disable x & y axis and show the rotated image
plt.axis('off')
plt.imshow(rotated_img)
plt.show()

# Save the rotated image to disk
plt.imsave("city_rotated.jpg", rotated_img)
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def show_image_with_caption(image, caption, filename):
    plt.axis('off')
    plt.imshow(image)
    plt.title(caption)
    plt.show()
    plt.imsave(filename, image)

# Read and convert the input image
img = cv2.cvtColor(cv2.imread("doggy.jpg"), cv2.COLOR_BGR2RGB)

# Show the original image
show_image_with_caption(img, "Original Image", "doggy_original.jpg")

# Get the image shape
rows, cols = img.shape[:2]

# Translation transformation 
M_translate = np.float32([[1, 0, 200], 
                          [0, 1, 200], 
                          [0, 0, 1]])
translated_img = cv2.warpPerspective(img, M_translate, (cols, rows))
show_image_with_caption(translated_img, "Translated Image", "doggy_translated.jpg")

# Scaling transformation
M_scale = np.float32([[1.5, 0, 0],
                      [0, 1.8, 0],
                      [0, 0, 1]])
scaled_img = cv2.warpPerspective(img, M_scale, (cols * 2, rows * 2))
show_image_with_caption(scaled_img, "Scaled Image", "doggy_scaled.jpg")

# Rotation transformation
angle = np.radians(10)
rotated_img = cv2.warpPerspective(img, np.float32([[np.cos(angle), -np.sin(angle), 0], [np.sin(angle), np.cos(angle), 0], [0, 0, 1]]), (cols, rows))
show_image_with_caption(rotated_img, "Rotated Image", "doggy_rotated.jpg")
```
